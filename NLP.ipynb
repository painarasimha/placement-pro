```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
#Tokenizing the entire file into words and sentences.
words = word_tokenize(text)
sentences = sent_tokenize(text)

print(words)
print(sentences)
```

```python
#Splitting the Sentences list into elements at '\n' separator
split_list = [item.split('\n') for item in sentences]

final_list = [item for sublist in split_list for item in sublist]
final_list
```

Using Regex Patterns to extract data from the PDF file

import re
# Join the lines into a single string
text = '\n'.join(final_list)

# Define regex patterns
name_pattern = r'^[A-Z][a-z]*\s[A-Z][a-z]*\s[A-Z][a-z]*'
phone_pattern = r'\+\d{2}\s\d{5}\s\d{5}'
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
cgpa_pattern = r'\b\d+\.\d{2}\b'
skills_pattern = r'Skills\b.*?((?:[A-Za-z]+,?\s?)+)'
achievements_pattern = r'Achievements\b(.*?)(?=Leadership\b|\b[A-Z]|Skills\b|\Z)'


# Match patterns in the text
name_match = re.search(name_pattern, text)
phone_numbers_match = re.findall(phone_pattern, text)
emails_match = re.findall(email_pattern, text)
cgpa_match = re.search(cgpa_pattern, text)
skills_match = re.search(skills_pattern, text, re.DOTALL)
achievements_match = re.search(achievements_pattern, text, re.DOTALL)

# Extract information
name = name_match.group() if name_match else None
phone_number = phone_numbers_match if phone_numbers_match else ''
email = emails_match if emails_match else ''
cgpa = cgpa_match.group() if cgpa_match else None
skills = skills_match.group(1).split(",") if skills_match else []
achievements = achievements_match.group(1).strip() if achievements_match else ''


# Print the extracted information
print("Name:", name)
print("Phone Number:", phone_number)
print("Email:", email)
print("CGPA:", cgpa)
print("Skills:", skills)
print("Achievements:", achievements)
